#### Total Sale by fat content  

```python
# we have to find the sales according to the fat content 
sales_by_fat = bk.groupby("Item Fat Content")['Sales'].sum()
print(f"sales by fat content is :",sales_by_fat)
```

```python
# now draw a pie chart 
plt.figure(figsize = (3,3))
plt.pie(sales_by_fat,labels= sales_by_fat.index,autopct = '%.1f%%',startangle = 90)
plt.title("sales by fat content")
plt.show
```

#### Total sale by item type

```python
Sales_by = bk.groupby('Item Type')['Sales'].sum().sort_values(ascending = False)
Sales_by
```

```python
# Now draw a graph 
plt.figure(figsize = (10,6))
bars = plt.bar(Sales_by.index,Sales_by.values)

plt.xticks(rotation = -90)
plt.xlabel('Item Type')
plt.ylabel('TotaL Sale')
plt.title('Total sale by item type')

# this function is used to giving the data labels on the bar chart 
for bar in bars:
    plt.text(bar.get_x()+bar.get_width()/2,bar.get_height(),
            f'{bar.get_height():,.0f}',ha = 'center',va = 'bottom',fontsize =8)

plt.tight_layout()
plt.show
```

#### Fat content by outlet for total sale 

```python
Sales_by_outlet = bk.groupby(['Outlet Location Type','Item Fat Content'])['Sales'].sum().unstack()
Sales_by_outlet = Sales_by_outlet[['Regular','Low Fat']]
```

```python
plt.Figure(figsize = (10,6))
ax = Sales_by_outlet.plot(kind ='bar')

plt.title('Outlet Tier By Fat Content')
plt.xlabel('outlet location tyer')
plt.ylabel('total sale')
```

#### Total sale by outlet estabilishement

```python
Sales_by_outlet_estabilishment = bk.groupby('Outlet Establishment Year')['Sales'].sum().sort_index()
Sales_by_outlet_estabilishment
```

```python
plt.figure(figsize = (10,6))
plt.plot(Sales_by_outlet_estabilishment.index,Sales_by_outlet_estabilishment.values,marker = 'o' , linestyle = '-')


# FOR SHOWING THE NUMBER OF THE SEAL

for x,y in zip (Sales_by_outlet_estabilishment.index,Sales_by_outlet_estabilishment.values):
    plt.text(x,y,f'{y:,.0f}',ha = 'left',va = 'bottom',fontsize = 8)


plt.title('Sales_by_outlet_estabilishment')
plt.xlabel('Establishment Year')
plt.ylabel('Total Sale')

plt.tight_layout()
plt.show()
```

#### Total sales by outlet size

```python
Sales_by_outlet_size = bk.groupby('Outlet Size')['Sales'].sum()
Sales_by_outlet_size
```

```python
plt.figure(figsize = (3,3))
plt.pie(Sales_by_outlet_size, labels = Sales_by_outlet_size.index,autopct = '%.1f%%',startangle = 90)
plt.title("sales by outlet size")
plt.tight_layout()
plt.show
```

#### Total sale by outlet location 

```python
Sales_by_outlet_location = bk.groupby('Outlet Location Type')['Sales'].sum().reset_index()
Sales_by_outlet_location = Sales_by_outlet_location.sort_values("Sales",ascending = False)

plt.figure(figsize = (8,3))
ax = sns.barplot(x = 'Sales',y = 'Outlet Location Type',data = Sales_by_outlet_location )
plt.title('Total Sales by outlet location')
plt.xlabel('Total sales')
plt.ylabel('Outlet Location Type')

plt.show()
```
